```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'stock_data'
# Filter data for the years 2020 to 2022
filtered_data = stock_data[(stock_data['date'] >= '2020-01-01') & (stock_data['date'] <= '2022-12-31')]

# Convert 'date' column to datetime format
filtered_data['date'] = pd.to_datetime(filtered_data['date'])

# Group data by year and month, and calculate mean for each month
monthly_data = filtered_data.groupby([filtered_data['date'].dt.year.rename('year'), filtered_data['date'].dt.month.rename('month')]).mean().reset_index()

# Create a new 'date' column from 'year' and 'month'
monthly_data['date'] = pd.to_datetime(monthly_data[['year', 'month']].assign(day=1))

# Plotting
plt.figure(figsize=(12, 8))

# Example: Plotting Open Price
plt.plot(monthly_data['date'], monthly_data['open'], label='Open Price')

# Example: Plotting Low and High Prices
plt.plot(monthly_data['date'], monthly_data['low'], label='Low Price')
plt.plot(monthly_data['date'], monthly_data['high'], label='High Price')

plt.title('Average Stock Prices by Month (2020-2022)')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)
plt.tight_layout()
plt.show()
```
