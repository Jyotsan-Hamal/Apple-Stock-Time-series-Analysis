```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../stock_datas/sanima.csv")
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.head()
```

```python
def calculate_close(row):
    return (row['high'] + row['low'] + row['ltp']) / 3
```

```python
# Apply the function to create the 'close' column
df['close'] = df.apply(calculate_close, axis=1)
```

```python
df.head()
```

```python
# Set 'date' column as index
df.set_index('date', inplace=True)
```

```python
df.tail()
```

```python
def ploto(fig ,name,df):
    plt.figure(figsize=fig)
    df[name].plot()
    plt.xlabel('years')
    plt.ylabel(f'{name}')
    plt.title(f"{name} data for Sanima data")
```

```python
plot((15,5),'close',df)
```

```python
plot('high')
```

```python
plot('open')
```
